```python
trip1_1=[]
trip1_2=[]
trip1_3=[]
trip4_2=[]
trip4_3=[]
trip_arrival =[]
for index, row in distance.iterrows():
    trip1_list = row["trip1"].split('.')
    trip1_1.append(trip1_list[0])
    trip1_2.append(trip1_list[1])
    trip1_3.append(trip1_list[2])
    trip4_list = row["trip4"].split('.')
    trip4_2.append(trip4_list[1])
    trip4_3.append(trip4_list[2])
    trip_arrival.append(int(row["arrival_time"].split(":")[0])*60*60 + int(row["arrival_time"].split(":")[1])*60 +int(row["arrival_time"].split(":")[2]))
```

```python
def get_nextanddistance(group):
    group.sort_values(['stop_sequence'],ascending=True)
    group["distance"] = -(group.shape_dist_traveled.diff(periods=-1))
    group["time_period"] = -(group.arrival_time.diff(periods=-1))
    group["previous_stopID"] = group.new_stop_id.shift()
    group["hour"] = group.departure_time/60/60
    return group
```

```python
def get_destinationstop(group):
    group.sort_values(['prog_number'],ascending=False)
    max_sequence_row = group.ix[group['prog_number'].idxmax()]
    group["route_end_stop"] = max_sequence_row.new_stop_id
    return group
def get_departurestop(group):
    group.sort_values(['prog_number'],ascending=True)
    max_sequence_row = group.ix[group['prog_number'].idxmin()]
    group["route_start_stop"] = max_sequence_row.new_stop_id
    return group
```

```python
def get_departuretime(group):
    group.sort_values(['prog_number'],ascending=True)
    min_sequence_row = group.ix[group['prog_number'].idxmin()]
    group["route_start_time"] = min_sequence_row.departure_time
    return group

def get_destinationtime(group):
    group.sort_values(['prog_number'],ascending=False)
    max_sequence_row = group.ix[group['prog_number'].idxmax()]
    group["route_end_time"] = max_sequence_row.departure_time
    return group
```

```python
distance = distance.groupby(["trip_id"]).apply(get_destinationstop)
distance = distance.groupby(["trip_id"]).apply(get_departurestop)
distance = distance.groupby(["trip_id"]).apply(get_departuretime)
distance = distance.groupby(["trip_id"]).apply(get_destinationtime)
```
