```python
import json
import re
import urllib.request
json_link = "https://data.dublinked.ie/cgi-bin/rtpi/busstopinformation?operator=bac&format=json"
```

### 1. acquire the stop data with the Dublin bus API

```python
json_data = urllib.request.urlopen(json_link)
data = json.load(json_data)
```

```python
data
```

```python
stops = [];
for stop_info in data["results"]:
    stop = {}
    try:
        stop["stop_id"] = stop_info["stopid"]
        stop["stop_name"]=stop_info["fullname"]
        stop["stop_lat"] = stop_info["latitude"]
        stop["stop_lon"] = stop_info["longitude"]
        stops.append(stop)
    except AttributeError:
        Print("Something wrong here")
    
```

```python
stops
```

```python
with open('stops.json','w') as outfile:
    json.dump(stops,outfile)
```

The issus is that many stops have the same fullname, which can not specify the exact stop.

### 1. acquire the stop data with STFC file with google

```python
json_file ="stops.json"
```

```python
json_data = open(json_file)
stops = json.load(json_data)
json_data.close()
```

```python
for stop in stops:
    try:
        stop["stop_id"] = re.search('([1-9][0-9]*\Z)', stop["stop_id"]).group(1)
    except AttributeError:
        Print("The results are not found")
```
