```python
def get_nextanddistance(group):
    group.sort_values(['stop_sequence'],ascending=True)
    group["distance"] = group.shape_dist_traveled.diff()
    group["previous_stopID"] = group.new_stop_id.shift()
    return group
```

```python
def get_destinationstop(group):
    group.sort_values(['stop_sequence'],ascending=False)
    max_sequence_row = group.ix[group['stop_sequence'].idxmax()]
    group["destination_stop"] = max_sequence_row.new_stop_id
    return group
def get_departurestop(group):
    group.sort_values(['stop_sequence'],ascending=True)
    max_sequence_row = group.ix[group['stop_sequence'].idxmin()]
    group["departure_stop"] = max_sequence_row.new_stop_id
    return group
```

```python
distance_1 = distance_1.groupby(["trip_id"]).apply(get_destinationstop)
distance_1 = distance_1.groupby(["trip_id"]).apply(get_departurestop)
```
